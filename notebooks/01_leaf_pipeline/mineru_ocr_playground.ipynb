# MinerU OCR Playground

这个 notebook 用来测试：`pdf/图片 -> MinerU -> markdown/json/txt`。

- 默认只写到 `scratch/mineru_runs/`
- 不会导入 `wrongbook`
- 不会改正式知识库
- 首次运行可能会下载较大的 OCR 模型


```python
import json
import subprocess
from pathlib import Path

PROJECT_ROOT = Path('/Users/xumuchi/Desktop/TeachAgent')
SCRIPT_PATH = PROJECT_ROOT / 'scripts' / 'run_mineru_extract.py'
OUTPUT_ROOT = PROJECT_ROOT / 'scratch' / 'mineru_runs'

INPUT_PATH = Path('/Users/xumuchi/Desktop/source/640.png')
RUN_NAME = 'demo_640_png'
BACKEND = 'pipeline'
LANG = 'ch'
METHOD = 'ocr'
ENABLE_FORMULA = True
ENABLE_TABLE = False

print({'script_exists': SCRIPT_PATH.exists(), 'input_exists': INPUT_PATH.exists(), 'output_root': str(OUTPUT_ROOT)})
```

```python
cmd = [
    'python3',
    str(SCRIPT_PATH),
    '--input', str(INPUT_PATH),
    '--run-name', RUN_NAME,
    '--backend', BACKEND,
    '--lang', LANG,
    '--method', METHOD,
    '--formula' if ENABLE_FORMULA else '--no-formula',
    '--table' if ENABLE_TABLE else '--no-table',
    '--force',
]

result = subprocess.run(cmd, capture_output=True, text=True)
print('returncode =', result.returncode)
print(result.stdout[:5000])
if result.returncode != 0:
    print('===== STDERR =====')
    print(result.stderr[:8000])
```

```python
run_dir = OUTPUT_ROOT / RUN_NAME
summary_path = run_dir / 'ocr_run_summary.json'
preview_path = run_dir / 'ocr_preview.txt'

summary = json.loads(summary_path.read_text(encoding='utf-8'))
preview_text = preview_path.read_text(encoding='utf-8')

print(json.dumps(summary, ensure_ascii=False, indent=2))
```
